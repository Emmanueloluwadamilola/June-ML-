```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn import metrics
from sklearn.linear_model import Lasso
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost.sklearn import XGBClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn

import lightgbm as lgb
from sklearn import metrics
```

```python
train.columns[1:-1]
```

```python
scaler = StandardScaler()
model = scaler.fit(train.drop(['target','id'],axis=1))

scaled_train = model.transform(train.drop(['target','id'],axis=1))
```

```python
train_new = pd.DataFrame(scaled_train, columns = train.columns[1:-1])
train_new.head()
```

```python
train['id']
```

```python
X = train_new.values
y = train['target'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
lm = LogisticRegression()
lm_model = lm.fit(X_train,y_train)
lm_prediction = lm_model.predict(X_test)
```

```python
lm_prediction
```

```python
def classifier():
    #logistic regression
    clf1= LogisticRegression()
    print("                       Logistic Regression")
    clf1.fit(X_train,y_train)
    lm_pred=clf1.predict(X_test)
    print(metrics.classification_report(y_test, lm_pred))
    print('\n')

    #Random Forest Classifier
    clf2= RandomForestClassifier()
    print("                       Random Forest Classifier")
    clf2.fit(X_train,y_train)
    RF_pred = clf2.predict(X_test)
    print(metrics.classification_report(y_test, RF_pred))
    print('\n')

    #XGBoost Classifier
    clf3= XGBClassifier(eval_metric='mlogloss')
    print("                       XGBoost Classifier")
    clf3.fit(X_train,y_train)
    XG_pred=clf3.predict(X_test)
    print(metrics.classification_report(y_test, XG_pred))
    print('\n')

    #AdaBoost Regression
    clf4= AdaBoostClassifier()
    print("                       AdaBoost Classifier")
    clf4.fit(X_train,y_train)
    Ada_pred=clf4.predict(X_test)
    print(metrics.classification_report(y_test, Ada_pred))
    print('\n')

  #Extra Trees Classifier
    clf5= ExtraTreesClassifier()
    print("                       Extra Trees Classifier")
    clf5.fit(X_train,y_train)
    ET_pred=clf5.predict(X_test)
    print(metrics.classification_report(y_test, ET_pred))
    print('\n')

  #Decision Tree Classifier
    clf6= DecisionTreeClassifier()
    print("                       Decision Tree Classifier")
    clf6.fit(X_train,y_train)
    DT_pred=clf6.predict(X_test)
    print(metrics.classification_report(y_test, DT_pred))
    print('\n')

  #LGBM Classifier
    clf7= lgb.LGBMClassifier()
    print("                       LGBM Classifier")
    clf7.fit(X_train,y_train)
    LGM_pred=clf7.predict(X_test)
    print(metrics.classification_report(y_test, LGM_pred))
    print('\n')



classifier()  
```

```python
error_rate = []

# Will take some time
for i in range(1,100):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
    
    
    
plt.figure(figsize=(20,20))
plt.plot(range(1,100),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```
